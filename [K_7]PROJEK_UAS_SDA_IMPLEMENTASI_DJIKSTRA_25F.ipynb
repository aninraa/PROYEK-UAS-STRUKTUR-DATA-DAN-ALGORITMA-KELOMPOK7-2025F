**IMPLEMENTASI ALGORITMA DIJKSTRA DALAM PENENTUAN RUTE EVAKUASI BENCANA KEBAKARAN PADA AREA PABRIK PT PROK CONVEYOR COMPONENTS**

Sumber: Azis, R., Nugrahadi, B., Devi, A. O. T., & Primasanti, Y. (2024). Perancangan Jalur Evakuasi Kebakaran PT. PCC dengan Metode Algoritma Dijkstra. Jurnal Teknik Mesin, Industri, Elektro dan Informatika, 3(3), 199-212. https://doi.org/10.55606/jtmei.v3i3.4208

```python
import heapq

class RuteEvakuasi:
    def __init__(self, graph_data):
        self.graph = graph_data                             # Menyimpan data graf berupa dictionary adjacency list

    def djikstra(self, start, target):
        dist = {node: float('inf') for node in self.graph}
        dist[start] = 0

        prev = {node: None for node in self.graph}

        pq = [(0, start)]

        while pq:
            d, u = heapq.heappop(pq)                        # heapq untuk mencari jarak terdekat

            if d > dist[u]:                                 # Jika jarak lebih besar dari jarak tersimpan berarti data lama dan tidak perlu diproses
                continue
            if u == target:                                 # Berhenti jika node target sudah ditemukan
                break

            for v, w in self.graph[u].items():
              # Menghitung alternatif jarak baru
                alt = dist[u] + w                           # Jarak node+bobot sisi

                if alt < dist[v]:                           # Jika jarak baru lebih pendek, update jarak dan simpan
                    dist[v] = alt
                    prev[v] = u
                    heapq.heappush(pq, (alt, v))            # Masukkan node ke pq agar kembali diproses

        return dist[target], prev                           # Untuk mengembalikan total jarak dan jejak rute


    def get_path(self, prev, target):
      # Untuk membentuk rute dari node tujuan
        path = []
        curr = target
        while curr is not None:                             # Menelusuri node sebelumnya sampai none
            path.append(curr)
            curr = prev[curr]
        return path[::-1]                                   # Balik urutan path agar menjadi start -> target



# Pengelompokan ruang di area gedung
graph = {
    # Gedung Area Depan(Titik Kumpul: X)
    "A": {"B": 10.8, "X": 15.7},
    "B": {"A": 10.8, "C": 9.6, "E": 8.2},
    "C": {"B": 9.6, "D": 13.2},
    "D": {"C": 13.2},
    "E": {"B": 8.2, "F": 3.6},
    "F": {"E": 3.6, "G": 4.2},
    "G": {"F": 4.2, "H": 8.7},
    "H": {"G": 8.7},
    "I": {"J": 4.5, "L": 6.2, "X": 16.3},
    "J": {"I": 4.5, "K": 8.2},
    "K": {"J": 8.2},
    "L": {"E": 14.8, "I": 6.2, "M": 11.2},
    "M": {"L": 11.2, "N": 16.2},
    "N": {"M": 16.2},
    "X": {"A": 15.7, "I": 16.3},
    # Gedung Area Belakang(Titik Kumpul: Y)
    "1": {"2": 11.2},
    "2": {"1": 11.2, "3": 11.3},
    "3": {"2": 11.3, "4": 11.3},
    "4": {"3": 11.3, "Y": 15.7},
    "5": {"6": 11.9},
    "6": {"5": 11.9, "7": 11.9},
    "7": {"6": 11.9, "8": 12},
    "8": {"7": 12, "Y": 11.6},
    "9": {"Y": 49.6},
    "Y": {"4": 15.7, "8": 11.6, "9": 49.6}
}

# Kelompok Node
tk_area_depan = ["A", "B", "C", "D", "E", "F", "G",
                    "H", "I", "J", "K", "L", "M", "N"]
tk_area_belakang = ["1", "2", "3", "4", "5",
                 "6", "7", "8", "9"]


if __name__ == "__main__":
    evakuasi = RuteEvakuasi(graph)  # Membuat objek sistem analisis dari class

    print("=== SISTEM ANALISIS JALUR EVAKUASI TERCEPAT ===")
    start_node = input("Masukkan Kode Lokasi Api: ").upper()  #Mengambil input huruf besar

    # Menentukan target titik kumpul otomatis berdasarkan wilayah gedung
    if start_node in tk_area_depan:
        target_node = "X"                 # Jika di area depan, diarahkan ke titik kumpul X
    elif start_node in tk_area_belakang:
        target_node = "Y"                 # Jika di area belakang, diarahkan ke titik kumpul Y
    else:
        target_node = None                # Jika kode ruang tidak terdaftar

    # Memproses rute jika lokasi target valid
    if target_node:
        jarak, jejak = evakuasi.djikstra(start_node, target_node) # Menghitung jarak terkecil dan mendapatkan rekam jejak alur

        rute = evakuasi.get_path(jejak, target_node)

        estimasi_detik = jarak / 1.33     # Menghitung estimasi waktu tempuh evakuasi

        # Mencetak hasil analisis
        print("\n===== RUTE TERCEPAT EVAKUASI =====")
        print(f"LOKASI KEBAKARAN\t: {start_node}")
        print(f"TITIK KUMPUL\t\t: {target_node}")
        print(f"RUTE TERECEPAT\t\t: {' -> '.join(rute)}") # Menggabungkan list dengan panah
        print(f"TOTAL JARAK\t\t: {jarak} meter")
        print(f"ESTIMASI WAKTU\t\t: {round(estimasi_detik, 2)} detik")
        print("="*40)
    else:
        print("\n[!] Maaf, ruangan tersebut tidak terdaftar di peta.")

```
